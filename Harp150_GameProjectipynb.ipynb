We have created an escape room. it combines two rooms, some distractions, and twists and turns along the way
Try and find the keys you need to escape!!

Escape Room



```python
count=0
count2=0
keycount=0
# these counts are used for when the user does a direction more than once

escape_room={"room1":
             {"up":"exit",
             "right":"room2",
             "south":"lamp",
             "left": "statue"},

             "room2":
             {"up":"curtain",
              "left":"room1",
              "right":"skeleton",
              "skeleton":"golden_key",
              "south":"glass box"}}
# here we created a dictionary of the directions, and their corresponding rooms/ objects 
# so we can remeber where everything is more effectively

import time 

def suspense(): # this creates a funtion called suspense
     time.sleep(1) # this waits a certain amount of seconds before printing the next code
     print("    . . .") # this prints ". . ." for added suspense
     time.sleep(1) # this waits a certain amount of seconds before printing the next code
# we created a suspense function to add a ". . ." and more time between phrases to make the game more suspensful
        
# introduction function for the beginning of the game that gives the title, premise, and outline of the game       
def introduction(): # this creates a function called introduction
    print('''
                          The House of Terror

...............................................................................

You've been thrown into a haunted house by your friends, and you need to find a way to escape!''') # print function
    
    suspense() # brings the suspense function
    print("to your north is a closed door") # print function
    suspense() # brings the suspense function
    print("to your right, you see an open door") # print function
    suspense() # brings the suspense function
    print("to the south, you see a dusty " + escape_room["room1"]["south"]) # this calls for the dictionary value from escape room --> room1 --> south direction
    suspense() # brings the suspense function
    print("and to your left you see a " + escape_room["room1"]["left"]) # print function that calls for a dictionary function from escape_room --> room1 --> left
    

    
inventory=[] # here we call for the inventory for when the player finds the key

introduction() # calls the introduction function when the player starts

while True: # this makes sure that the player can keep putting in different directions after the if elif else loop is completed
    suspense() # brings the suspense function
    room1_input=input("Ok.....which direction will you go now? north, left, right, or south? ").lower() # this will print every time the loop runs, 
    # allowing the player to input whatever direction desired
    
    if room1_input=="north": #the ultimate goal of escaping 
        suspense() # brings the suspense function
        print("this door is locked. Try and find something that will open it! ") # print function
    elif room1_input=="south" and count==0 : # this is if the player inputs south for the first time
        suspense()
        lamp_input=input("Do you want to turn it on? this could be a clue... or not. (yes or no) ").lower() # this asks for a yes or no from the player, and makes the input lowercase
        if lamp_input=="yes": # this is for if the player chooses not the turn the lamp on
            suspense() # brings the suspense function
            print("the lamp is now on. HAHA you have been tricked. It doesn't do anything. ") # print function
            count+= 1 # this adds to the count, so if the user comes back to the lamp, it goes to a different elif statement

        elif lamp_input=="no": # this is for if the player chooses not the turn the lamp on
            suspense() # brings the suspense function
            print("oh well. suit yourself") # print function
        else: # this is for when the player inputs something that is not an option
            print("""
      
      . . .
      
      
      we're not that bright are we... choose one of the      
      choices provided... 

      let's go back.
      
      """) # print function

    elif room1_input=="south" and count>=0: # this is for when the player chooses to go south for the second time
        suspense()
        print("you've messed with the " + escape_room["room1"]["south"] + " already... maybe try other things") # print function, calls for a value from the dictionary

    elif room1_input== "left": # this loop is for when the player chooses to go left
        suspense() # brings the suspense function
        print("the " + escape_room["room1"]["left"] + " stares at you...") # print function
        suspense() # brings the suspense function
        print("well... that wasn't helpful") # print function
    elif room1_input=="right":  # this is for when the player chooses to go right
        suspense() # brings the suspense function
        print("you are now in " + escape_room["room1"]["right"]) # print function
        suspense() # brings the suspense function
        print("its very dark in here, but there is a crack of light in front of you . . . ") # print function

        
        curtain=input("you move forward and feel a curtain swinging. It draws you in and you walk up. Do you want to move it aside? (yes or no) ").lower()
        # this will be prompted when player chooses to move to room 2, and asks for an input, and makes it all lowercase
        if curtain=="yes": # this loops is played if the player inputs yes
            suspense() # brings the suspense function
        elif curtain=="no": # this loop is played if the player inputs no
            suspense() # brings the suspense function
            print("a gust a wind blew the curtain open anyway!") # print function
        else: # this is if the player inputs somehthing thats not yes or no
            suspense() # brings the suspense function
            print("that wasn't a choice but a gust a wind blew the curtain open") # print function
        break    # this breaks the if statement for room 1           
    else: # this else statement is in case the player chooses something that is not an option.
        print("""
      
      . . .
      
      
      we're not that bright are we... choose a direction from the choices provided
      
      """) # print function
        
while True: 
    suspense()
    print("""the room is well lit now! you see a """ + escape_room["room2"]["south"] + """ to the south of you, room1 to your left, and to your right there's a skeleton!""" )  #introduction to the choices
    suspense()
    room2_input=input("Which direction do you want to go? (south, left, or right?) ").lower() #asks for user input on direction and makes it lowercase
    if room2_input=="right" and keycount==0:
        suspense()
        print("you are now at the " + escape_room["room2"]["right"] + ", and a " + escape_room["room2"]["skeleton"] + " is now in your posession!") #if user types right it will print this 
        inventory.append("golden_key") #adds golden_key to inventory
        keycount+=1 #adds count 
    elif room2_input =="right" and keycount>=0: #if they try to go right multiple times it would print the print statement below
        suspense()
        print("what a tragedy. You collected a golden key... there is nothing more to see here. ")
    elif room2_input=="south" and count2==0:
        suspense()
        print("you are now at the " + escape_room["room2"]["south"]+ "! maybe something is inside it. Do you want to open it? (yes or no?)")
        glass_input=input().lower() #lowercases the input
        if glass_input=="yes": 
            suspense()
            print("You fished out a piece of paper... look for the key and you will succeed") #hint 
            count2+=1 #adds 1 to count2
        elif glass_input=="no":
            suspense()
            print("ok. keep playing")
        else: # this is for when the player inputs something that is not an option
            print("""
      
      . . .
      
      
      we're not that bright are we... choose one of the      
      choices provided... 

      let's go back.
      
      """) # print function
    elif room2_input=="south" and count2>=0:
        print("nothing much to see here... you looked here already")
        #when they already went south and selected a yes or a no, they will recieve this print statement instead of the output from yes or no
      
    elif room2_input=="left":
        suspense()
        print("you are now back in room1. You walk towards the locked door, hoping that this time it will open")
        suspense()
        if "golden_key" in inventory:
            print("the door swings open and you have escaped! congrats")
            break #stops the program, game is over!
        else:
        print("uh oh, the door is still locked. try looking for something that will open it!") #when you attempt to go back to the room without the key
        suspense()
        print("you are now back in room2")
    else: # this else statement is in case the player chooses something that is not an option.
        print("""
      
      . . .
      
      
      we're not that bright are we... choose a direction from the choices provided
      
      """) # print function   



#outside references:
#https://github.com/Ninedeadeyes/15-mini-python-games-tutorial-series/blob/master/H)The%20Haunted%20House.py
```
