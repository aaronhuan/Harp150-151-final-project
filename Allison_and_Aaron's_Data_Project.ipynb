This project is an analysis on a data set of the world's deaths from 1990 to 2019 that we found in Kaggle.
We chose this dataset because it related to both our majors. We were able to look into and compare the deaths from different diseases between the world and a country, we picked Afganistan. 

```python
import csv
with open("annual_deaths_by_causes.csv") as deaths_data: # opens the csv file downloaded
    deaths=csv.reader(deaths_data) # using function from csv module called reader - gives a parameter of roblox data
    deaths_rows=[row for row in deaths]
deaths_rows # calls the csv data file to show
```

```python
import pandas as pd # imports the pandas library, calling it pd to make it easier
deaths=pd.read_csv("annual_deaths_by_causes.csv") # calls the csv file "deaths" to make it easier to call
deaths # calls the data set and shows it clearly below
```

```python
avg_deaths_malaria_world= deaths.groupby("year")["malaria"].mean() # creates a grouped data set from the larger one, only showing how mnay people dies from malaria each year
avg_deaths_malaria_world # calls the grouped data set
```

```python
deaths_afg = deaths.copy() # makes a copy of deaths to use

deaths_afg.drop(deaths_afg[(deaths_afg["country"] != "Afghanistan")].index, inplace = True) # this drops all countries from the data set except afghanistan
deaths_afg # calls the afghanistan deaths data set
```

```python
afg_deaths_malaria= deaths_afg.groupby("year")["malaria"].mean() # groups how man people in afghanistan died from malaria each year 
afg_deaths_malaria # calls the data set
```

```python
import matplotlib.pyplot as plt # calls pyplot from matplot lib as plt
import numpy as np # calls numpy as np


labels = deaths_afg["year"]
x = np.arange(len(labels))  # the label locations
width = .6  # the width of the bars

fig, ax = plt.subplots() #makes space for graph
rects1 = ax.bar(x - width/2, avg_deaths_malaria_world, width, label='World Deaths') #labels blue bar graph in legend
rects2 = ax.bar(x + width/2, afg_deaths_malaria, width, label='Afghanistan')#labels orange bar graph in legend

ax.set_ylabel('Deaths') #labels the y-axis as "Deaths"
ax.set_xlabel("Years") #labels the x-axis as "Years"
ax.set_title('Deaths from malaria: world vs afg (1990-2019)') #sets title 
ax.set_xticks(x, labels, rotation='vertical') #sets x-axis labels(the years) in a vertical direction to see more clearly
ax.legend() #shows legend

fig.tight_layout() #makes sure nothing gets cut off

plt.show() #shows graph
```

```python
afg_deaths_alz= deaths_afg.groupby("year")["alzheimer's_diesease"].mean() # groups together how many people in afghanistan died from alzheimer's each year
afg_deaths_alz # calls the data set
```

```python
afg_deaths_par= deaths_afg.groupby("year")["parkinson's_disease"].mean() # groups together how many people in afghanistan died from parkinson's each year
afg_deaths_par # calls the data set
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = deaths_afg["year"]
x = np.arange(len(labels))  # the label locations
width = .6  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, afg_deaths_alz, width, label="Alzheimer's")  #plots and labels the blue bar 
rects2 = ax.bar(x + width/2, afg_deaths_malaria, width, label="Parkinson's") #plots and labels the orange bar 

ax.set_ylabel("Deaths") #sets y-axis as Deaths 
ax.set_xlabel("Years")  #sets x-axis as Years 
ax.set_title("Deaths in Afghanistan each year from Alzeimer's and Parkinsons") #creates the title
ax.set_xticks(x, labels, rotation='vertical') #makes x-axis labels (the years) vertical 
ax.legend() #shows legend

fig.tight_layout() #makes sure nothing gets cut off

plt.show()
```

```python
#world deaths by Alzheimer in 2019
avg_alz_2019 = deaths.copy() # copies the csv
avg_alz_2019.drop(avg_alz_2019[(avg_alz_2019["year"] != 2019)].index, inplace = True) #drops data that isn't from 2019
avg_alz_2019= avg_alz_2019.groupby("year")["alzheimer's_diesease"].mean() #gives the mean amount of alzheimer deaths 
avg_alz_2019 #shows mean
```

```python
#world deaths by Parkinson's in 2019
avg_par_2019 = deaths.copy()
avg_par_2019.drop(avg_par_2019[(avg_par_2019["year"] != 2019)].index, inplace = True) #drops data that isn't from 2019
avg_par_2019= avg_par_2019.groupby("year")["parkinson's_disease"].mean() #gives the mean amount of Parkinson's deaths 
avg_par_2019 #shows mean
```

```python
#Afganistan deaths by Alzheimer in 2019
afg_alz_2019 = deaths.copy()
afg_alz_2019.drop(afg_alz_2019[(afg_alz_2019["year"] != 2019)].index, inplace = True) #drops data that isn't from 2019
afg_alz_2019.drop(afg_alz_2019[(afg_alz_2019["country"] != "Afghanistan")].index, inplace = True) #drops data that isn't from Afghanistan
afg_alz_2019= afg_alz_2019.groupby("year")["alzheimer's_diesease"].mean()#gives the mean amount of alzheimer deaths 
afg_alz_2019 #shows mean
```

```python
#Afganistan deaths by Parkinson's in 2019
afg_par_2019 = deaths.copy()
afg_par_2019.drop(afg_par_2019[(afg_par_2019["year"] != 2019)].index, inplace = True) #drops data that isn't from 2019
afg_par_2019.drop(afg_par_2019[(afg_par_2019["country"] != "Afghanistan")].index, inplace = True) #drops data that isn't from Afghanistan
afg_par_2019= afg_par_2019.groupby("year")["parkinson's_disease"].mean()#gives the mean amount of Parkinson's deaths
afg_par_2019 #shows mean
```

```python
#pie chart world alz vs world par (Alzheimers and Parkinsons)

class pie_carts:
        
        #pie chart comparing world alz to park
    def world_alz_vs_park_in_2019(): #sets funciton
        labels = "World Alzheimer", "World Parkinson's" #labels
        numbers= [int(avg_alz_2019), int(avg_par_2019)] #defines variable numbers 
        explode = (0, .1) #seperates pie chart sections 

        fig1, ax1 = plt.subplots()
        plt.pie (numbers, explode = explode, labels = labels,
                autopct = "%1.1f%%", shadow = False, startangle = 90) #defines parameters for the chart 
        plt.axis("equal") #makes it look like a circle
        plt.title("World Deaths from Alzheimer's and Parkinsons in 2019") #creates the title

        plt.show
    world_alz_vs_park_in_2019()
           
        #pie chart comparing afghan alz to world alz
    def world_alz_vs_afg_alz_in_2019(): # sets function
        labels = "AFG Alzheimer", "World Alzheimer" #labels
        numbers= [int(afg_alz_2019), int(avg_alz_2019)] #defines variable numbers 
        explode = (0, .4)#seperates pie chart sections 

        fig1, ax1 = plt.subplots()
        plt.pie (numbers, explode = explode, labels = labels,
                autopct = "%1.1f%%", shadow = False, startangle = 120) #defines parameters for the chart 
        plt.axis("equal")#makes it look like a circle
        plt.title("World vs Afganistan deaths from Alzheimer's in 2019") #creates the title

        plt.show
    world_alz_vs_afg_alz_in_2019()
            
        #pie chart comparing afghan par to world par
    def world_vs_afg_park_in_2019():# sets function
        labels = "World Parkison's", "Afg Parkinson's" #labels
        numbers= [int(avg_par_2019), int(afg_par_2019)]#defines variable numbers 
        explode = (0, .1)#seperates pie chart sections 

        fig1, ax1 = plt.subplots()
        plt.pie (numbers, explode = explode, labels = labels,
                autopct = "%1.1f%%", shadow = False, startangle = 90)#defines parameters for the chart 
        plt.axis("equal")#makes it look like a circle
        plt.title("World vs Afganistan deaths from Parksinson's in 2019") #creates the title

        plt.show
    world_vs_afg_park_in_2019()
```

```python
# counts each year in the data set (shows that each year is not featured the same amount)
year_deaths = {}
for i in deaths ["year"]:
    if i in year_deaths:
        year_deaths[i] +=1
    else:
        year_deaths[i] = 1 
year_deaths
```

```python
liver_dis = deaths.copy() #makes copy of csv data set
liver_dis.drop(liver_dis[(liver_dis["country"] != "Afghanistan")].index, inplace = True) #drops data that isn't under Afganistan
liver_dis= liver_dis.groupby("year")["chronic_liver_diseases"].mean() #groups mean deaths by liver disease by year
liver_dis
```

```python
alcohol_dis = deaths.copy() #makes copy of csv data set
alcohol_dis.drop(alcohol_dis[(alcohol_dis["country"] != "Afghanistan")].index, inplace = True) #drops data that isn't under Afganistan
alcohol_dis= alcohol_dis.groupby("year")["alcohol_use_disorders"].mean()#groups mean deaths by Alcohol use by year
alcohol_dis
```

```python
import numpy as np
import matplotlib.pyplot as plt

y1=np.array(liver_dis) #plots deaths by liver disorder
y2=np.array(alcohol_dis) #plots death by alcohol use
plt.plot(y1, color='blue', label='alcohol disorders') #labels and colors line graph of alcohol use
plt.plot(y2,color='orange',label='liver disorders') #labels and colors line graph of Liver disorders
plt.xlabel("Years") #labels x-axis with "years"
plt.ylabel("Deaths") #labels y-axis with "Deaths"
plt.title("Deaths from Liver Disorders vs Alcohol Disorders") #labels title 

plt.legend() #shows legend
plt.show()#shows graph
```
